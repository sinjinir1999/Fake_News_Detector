FAKE NEWS DETECTOR | A SYSTEM TO IDENTIFY UNRELIABLE NEWS ARTICLES

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train.csv')
```

```python
train.head()
```

```python
train=train.dropna()
```

```python
x=train.drop('label',axis=1)
```

```python
x.reset_index(inplace=True)
```

```python
y=train['label']
```

```python
voc_size=10000
```

```python
x.shape
```

```python
y.shape
```

DATA PREPROCESSING

```python
import nltk
import re
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
corpus=[]
```

```python
for i in range(0,len(x)):
  review=re.sub('[^a-zA-Z]',' ',x['title'][i])
  review=review.lower()
  review=review.split()
  
  review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

WORD EMBEDDING REPRESENTATION

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.text import one_hot
```

```python
rep=[one_hot(i,voc_size) for i in corpus]
```

```python
rep
```

```python
sent_length=50
```

```python
embedded_docs=pad_sequences(rep,padding='pre',maxlen=sent_length)
```

```python
embedded_docs
```

```python
dim=50
```

LSTM | LONG SHORT TERM MEMORY MODEL

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
x_data=np.array(embedded_docs)
y_data=np.array(y)
```

```python
x_data.shape
```

```python
y_data.shape
```

SPLITTING OF DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.30,random_state=42)
```

MODEL TRAINING

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=30,batch_size=64)
```

PERFORMANCE METRICS AND ACCURACY

```python
y_pred=model.predict_classes(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
score=accuracy_score(y_test,y_pred)
```

```python
score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```
