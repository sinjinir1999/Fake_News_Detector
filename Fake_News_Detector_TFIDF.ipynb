FAKE NEWS DETECTOR | A SYSTEM TO IDENTIFY UNRELIABLE NEWS ARTICLES

```python
import pandas as pd
import nltk
import re
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
train=pd.read_csv('train.csv')
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train=train.dropna()
```

```python
train.isnull().sum()
```

```python
data=train.copy()
```

```python
data.reset_index(inplace=True)
```

```python
data=data.drop('label',axis=1)
```

```python
data.shape
```

```python
data.head()
```

TEXT PREPROCESSING | CLEANING THE DATA


```python

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
wnl=WordNetLemmatizer()
```

```python
corpus=[]
```

```python
for i in range(0,len(data)):
  review=re.sub('[^a-zA-Z]',' ',data['title'][i])
  review=review.lower()
  review=review.split()
  
  review = [wnl.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

TFIDF | Term Frequency and Inverse Document Frequency Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer(max_features=6000,ngram_range=(1,3))
```

```python
X=tf.fit_transform(corpus).toarray()
```

```python
Y=train['label']
```

```python
X.shape
```

```python
Y.shape
```

SPLITTING OF DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
tf.get_feature_names()
```

```python
tf.get_params()
```

```python
df = pd.DataFrame(x_train, columns=tf.get_feature_names())
```

```python
df.tail()
```

MODEL TRAINING

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
classifier.fit(x_train,y_train)
```

PREDICTION AND EVALUATION

```python
from sklearn import metrics
```

```python
y_pred=classifier.predict(x_test)
```

```python
score=metrics.accuracy_score(y_test,y_pred)
```
